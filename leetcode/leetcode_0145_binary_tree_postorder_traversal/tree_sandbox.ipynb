```python
from typing import Optional

from draw_tree import draw_tree
from tree import TreeNode, breadth_first_build
```

```python
t = breadth_first_build([1, 2, 3, 4, 5, 6, 7])
draw_tree(t)
```

```python
def post_order_recursive(root: Optional[TreeNode], visit):
    if root is None:
        return

    post_order_recursive(root.left, visit)
    post_order_recursive(root.right, visit)
    visit(root)
```

```python
post_order_recursive(t, print)
```

```python
def post_order(root: Optional[TreeNode], visit):
    stack = [(root, False)]
    while stack:
        node, ready = stack.pop()
        if node is None:
            continue
        if ready:
            visit(node)
        else:
            stack.append((node, True))
            stack.append((node.right, False))
            stack.append((node.left, False))
```

```python
post_order(t, print)
```
